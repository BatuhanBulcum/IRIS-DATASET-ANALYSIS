Let's define the functions.

```python
def Mean(u):
  mean = sum(u)/len(u)
  return mean
def Standard_Deviatoin(u):
  total = 0
  for i in u:
    total += ((Mean(u)-i) ** 2)
  return sqrt(total/(len(u)-1))
def Maximum(u):
  maximum=0
  for i in u:
    if i>maximum:
        maximum=i
  return maximum
def Minimum(u):
  minimum=Maximum(u)
  for i in u:
    if i<minimum:
        minimum=i
  return minimum
```

## **Covariance**

```python
def Coveriance(u,v):
  total=0
  for i,j in zip(u,v):
    total+=(Mean(u)-i)*(Mean(v)-j)
  return total/(len(u)-1)
```

```python
Coveriance(data['sepal_length'],data['sepal_length'])
```

```python
data.cov()
```

```python
Coveriance(setosa['sepal_length'],setosa['sepal_length'])
```

## **Correlation**

```python
def Correlation(u,v): 
  return Coveriance(u,v)/(Standard_Deviatoin(u)*Standard_Deviatoin(v))
```

```python
Correlation(data['sepal_length'],data['sepal_length'])
```
